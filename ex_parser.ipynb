# Example Election Results Scraper (Virginia)

Note: To run this code, scroll down and run the html_string = cell first
The election results are on this website: https://historical.elections.virginia.gov/elections/search/year_from:2017/year_to:2017/office_id:8/stage:General



```python
#Use Beautiful Soup
soup = BeautifulSoup(html_string)

#Take a look at the output
print(soup)
```

```python
#Iterate over the "a" elements that contain a link, split the link to just grab the ID

election_id_list = []
count = 0
for val in soup.find_all("a",href=True):
    if "elections/view" in val["href"]:
        election_id_list.append(val["href"].split("/")[3])
        count+=1
print(count)
```

```python
#From looking at the website, I noticed the pattern of the direct download links.
#Using the IDs from above, navigate to the direct download page and save it

for race in election_id_list:
    url = "https://historical.elections.virginia.gov/elections/download/"+race+"/precincts_include:1/"
    myfile = requests.get(url)
    open(myfile.headers['Content-disposition'][21:], 'wb').write(myfile.content)
```
